***Exploratory Data Analysis (EDA)***

```python
# استيراد المكتبات اللازمة
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

***Load Data***

```python
# تحديد مسار الملف
file_path = '~/code/Alanoudis/food-delivery-rec/data/updated_data/orders_sg10k.txt'

# قراءة الملف باستخدام pandas
data = pd.read_csv(file_path)

# عرض أول 5 صفوف من البيانات
data.head()
```

```python
# معرفة معلومات عامة عن البيانات
data_info = data.info()

# عرض بعض الإحصائيات الوصفية للبيانات
data_description = data.describe()

# طباعة النتائج
print(data_info)
print("\n")
print(data_description)
```

***Inspector data***

```python
# التحقق من القيم المفقودة في البيانات
missing_values = data.isnull().sum()

# طباعة عدد القيم المفقودة لكل عمود
print(missing_values)
```

***(Categorical Data Analysis)***

```python
# عرض القيم الفريدة في الأعمدة النوعية
categorical_columns = ['customer_id', 'geohash', 'vendor_id', 'product_id', 'day_of_week', 'order_time', 'order_day']
for col in categorical_columns:
    print(f"unique value {col}:")
    print(data[col].unique())
    print("\n")
```

***(Numerical Data Analysis)***

```python
# عرض الإحصائيات الوصفية للأعمدة الرقمية
numerical_data = data.describe()

# طباعة الإحصائيات الوصفية
print(numerical_data)
```

***Data Distribution Analysis using Visualizations***

```python
# Histogram for the distribution of orders by day of the week
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='day_of_week', palette='viridis')
plt.title('Order Distribution by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
plt.show()
```

***Plotting the Relationship Between Order Time and Day of the Week***

```python
# Convert order time to just the hour
data['order_hour'] = pd.to_datetime(data['order_time'], format='%H:%M:%S').dt.hour

# Plot the relationship between order time and day of the week
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='day_of_week', hue='order_hour', palette='coolwarm')
plt.title('Relationship Between Order Time and Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
plt.legend(title='Order Hour', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

***Scatter Plot Between Order Hour and Number of Orders***

```python
# Scatter plot between order hour and number of orders
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='order_hour', y=data.groupby('order_hour')['order_id'].transform('count'), palette='viridis')
plt.title('Scatter Plot Between Order Hour and Number of Orders')
plt.xlabel('Order Hour')
plt.ylabel('Number of Orders')
plt.show()
```

***Analyzing the Relationship Between Orders and Geographical Location (Geohash)***

```python
# Plotting the distribution of orders by geographical location
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='geohash', palette='Set2')
plt.title('Distribution of Orders by Geographical Location')
plt.xlabel('Geographical Location (Geohash)')
plt.ylabel('Number of Orders')
plt.xticks(rotation=90)  # Rotate labels for better alignment
plt.show()
```

***Plotting the Interactive Pairplot for Distributions***

```python
# Selecting only the numerical columns
numerical_columns = ['order_id', 'order_hour']

# Plotting the interactive pairplot
sns.pairplot(data[numerical_columns])
plt.title('Pairplot Between Numerical Columns')
plt.show()
```

***Analyzing Time Trends of Orders***

```python
# Convert the 'order_day' column to string, remove the word "days", and then convert it to an integer representing days
data['order_day'] = data['order_day'].astype(str).str.replace(' days', '').astype(int)

# Calculate the number of orders for each day
daily_orders = data.groupby('order_day').size()

# Plot the time trends
plt.figure(figsize=(12, 6))
daily_orders.plot(kind='line', color='blue')
plt.title('Time Trends of Orders Across Days')
plt.xlabel('Day')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)  # Rotate labels on the x-axis
plt.grid(True)
plt.show()
```

***Analyzing the Relationship Between Orders by Vendor ID***

***Displaying Only the Top 10 Vendors***

```python
# Calculate the number of orders by vendor_id
vendor_counts = data['vendor_id'].value_counts()

# Display the top 10 vendors only
top_10_vendors = vendor_counts.head(10)

# Plot the distribution of orders by the top 10 vendor_ids
plt.figure(figsize=(12, 6))
sns.countplot(data=data[data['vendor_id'].isin(top_10_vendors.index)], x='vendor_id', palette='Set2')
plt.title('Distribution of Orders by Top 10 Vendor IDs')
plt.xlabel('Vendor ID')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)  # Rotate labels on the x-axis
plt.show()
```

***Analyzing the Relationship Between Day of the Week and Order ID***

```python
# Plotting the distribution of orders by day of the week
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='day_of_week', palette='Blues')
plt.title('Distribution of Orders by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
plt.show()
```

***# Analyzing the Relationship Between Order Time and Order Day***

```python
# Convert order time to just the hour
data['order_hour'] = pd.to_datetime(data['order_time'], format='%H:%M:%S').dt.hour

# Plot the number of orders by hour across days
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='order_hour', hue='order_day', palette='coolwarm')
plt.title('Number of Orders by Hour Across Days')
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.show()
```

***Analyzing the Relationship Between Customer ID and Order ID***


```python
# Calculate the number of orders for each customer
customer_orders = data['customer_id'].value_counts()

# Display the top 10 customers by number of orders
plt.figure(figsize=(12, 6))
sns.barplot(x=customer_orders.head(10).index, y=customer_orders.head(10).values, palette='Set1')
plt.title('Top 10 Customers by Number of Orders')
plt.xlabel('Customer ID')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```
